```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def read_data(filename):
    """
    Чтение данных из файла. Ожидается, что файл содержит строки вида "h,sigma".
    Возвращает два списка: h_values и sigma_values.
    """
    h_values = []
    sigma_values = []

    with open(filename, 'r') as file:
        for line in file:
            h, sigma = line.strip().split(',')
            h_values.append(abs(float(h)))
            sigma_values.append(abs(float(sigma)))

    return np.array(h_values), np.array(sigma_values)

def plot_data(h_values, sigma_values):
    """
    Построение графика log(h) vs log(sigma) с точками и соединяющими их линиями.
    """
    plt.figure(figsize=(8, 6))

    result = stats.linregress(np.log10(h_values)[0:4], np.log10(sigma_values)[0:4])
    print("коэффициент наклона - ", result.slope, result.stderr, result.intercept )
     
   
    plt.scatter(np.log10(h_values), np.log10(sigma_values), color='blue', label='Data')
    plt.plot(np.log10(h_values), np.log10(sigma_values), color='blue', linestyle='-', alpha=0.5)  # alpha для прозрачности линии

    plt.plot(np.log10(h_values)[0:6], result.slope*np.log10(h_values)[0:6]+result.intercept, color='red', linestyle='--', label='y='+str(round(result.slope, 1))+'x'+' - порядок аппроксимации')
    
    plt.xlabel('log10(h)')
    plt.ylabel('log10(sigma)')
    plt.title('log10(h) vs log10(sigma)')
    plt.grid(True)
    plt.legend()

    # Показываем график
    plt.show()
```

```python
h_values, sigma_values = read_data('3')

plot_data(h_values, sigma_values)
```

```python
h_values, sigma_values = read_data('4')
plot_data(h_values, sigma_values)
```
